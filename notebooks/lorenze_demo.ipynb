# Lorenz butterfly demo

This notebook demonstrates how use a locally installed module for testing and development.

```python
# Import the class
from lorenze import LorenzButterfly
```

```python
# Generate data
lorenz = LorenzButterfly(sigma=10.0, rho=28.0, beta=8.0 / 3.0, steps=5000)
data = lorenz.generate_data()
data.head()
```

```python
# Plot the butterfly
lorenz.plot(color="tab:purple", linewidth=0.6)
```
